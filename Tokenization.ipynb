## Tokenization

### Prefix: Character(s) at the beggining $ ("
### Suffix: Character(s) at the end km ) , . ! "
### Infix: Character(s) in between - --/ ...
### Exception: Special-case rule to split a string into several tokens prevent a token form being split when punctiation rules are applied let's U.S.


## Token Methods

```python
import spacy
nlp = spacy.load("en_core_web_sm")
```

```python
mystring = '"We\'re moving to L.A.!"'
```

```python
print(mystring)
```

```python
doc = nlp(mystring)
```

```python
for token in doc:
    print(token.text)
```

```python
doc2 = nlp(u"We're here to help! Send snail-mail, email support@oursite.com or visit us at http://www.oursite.com")
```

```python
doc8 = nlp(u"Apple to build a Hong Kong factory for $6 million")
```

```python
for token in doc8:
    print(token.text, end = " | ")
    
```

```python
for entity in doc8.ents:
    print(entity," | ",entity.label_," | ",str(spacy.explain(entity.label_)), "\n")
```

```python
doc9 = nlp(u"Autonomous cars shift insurance liability towards manufacturers.")
```

```python
for chunk in doc9.noun_chunks:
    print(chunk)
```

## Tokenization Visualized

```python
from spacy import displacy
```

```python
doc10 =  nlp(u"Apple is going to build a U.K. factory for $6 million.")
```

```python
displacy.render(doc10,style = "dep",jupyter = True, options= {"distance":110})
```

```python
doc11 = nlp(u"Over the last quarter Apple sold nearly 20 thousand iPods for a profit of  $6 million")
```

```python
displacy.render(doc11,style = "ent",jupyter = True)
```

#### dep -> dependencies
#### ent -> entitites

```python
# using server
displacy.serve(doc11,style="dep")
```
